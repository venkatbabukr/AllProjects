```python
import numpy as np
import matplotlib.pyplot as plt

# Many functions
def sqrt_quadratic_x(x):
  return np.sqrt((x - 2)**2)

def nlogn(x):
  return x * np.log(x)

def simple_line(x):
 return 2*x + 5

def parabolic_with_displacement(x):
  return (x - 5)**2 + 10

def cubic_with_displacement(x):
  return (x - 5)**3 + 10

x = np.arange(-20, 20, 0.01)

def setup_plot():
  # 1. Use a base style
  plt.style.use('bmh')

  fig, ax = plt.subplots(figsize=(10, 10))
  # Set max y limit for now to 50
  ax.set_ylim(-50, 50)
  # 2. Move Spines to the center (origin)
  plt.setp(ax.spines['left'], position='zero', color='black', linewidth=2, zorder=0)
  plt.setp(ax.spines['bottom'], position='zero', color='black', linewidth=2, zorder=0)
  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')

def show_plot(plot_title: str):
  plt.title(plot_title)
  plt.grid(True)
  # plt.xlabel('X value')
  # plt.ylabel('Y value')
  plt.legend()
  plt.show()
```

```python
# Sublinear and linear functions plots

setup_plot()

y = np.log10(x)
plt.plot(x, y, label = r'Lograthmic: $log_{10}(x)$')

y = np.log(x)
plt.plot(x, y, label = r'Lograthmic: ln(x)')

y = np.sqrt(x)
plt.plot(x, y, label = r'Sqrt: $\sqrt{x}$')

y = sqrt_quadratic_x(x)
plt.plot(x, y, label = r'Sqrt quadratic: $\sqrt{(x-2)^2}$')

y = simple_line(x)
plt.plot(x, y, label = r'Linear: mx + c')

y = nlogn(x)
plt.plot(x, y, label = r'Linear Lograthmic: xln(x)')

show_plot('Sublinear functions')
```

```python
# Quadratic plots / Parabolas
setup_plot()

y = np.abs(x)
plt.plot(x, y, label = r'Absolute value: |x|')

y = np.square(x)
plt.plot(x, y, label = r'Parabola: $x^2$')

y = parabolic_with_displacement(x)
plt.plot(x, y, label = r'Parabola with displacement')


show_plot('Quadratic functions/Parabolas')
```

```python
# Cubic functions
setup_plot()

y = np.power(x, 3)
plt.plot(x, y, label = r'Cubic: $x^3$')

y = cubic_with_displacement(x)
plt.plot(x, y, label = r'Cubic with displacement')

show_plot('Cubic functions')
```
