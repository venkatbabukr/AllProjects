```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dummy 1D data points belonging to two classes
class_A = np.array([-2, -1.5, -1, 1, 1.5, 2]) # yi = +1
class_B = np.array([-4, -3, -2.5,  2.5, 3, 4]) #yi = -1


# Plot the original 1D points
# plt.figure(figsize=(12, 6))
plt.scatter(class_A, np.zeros_like(class_A), color='red', label='Class Pos (+1)', marker='o')
plt.scatter(class_B, np.zeros_like(class_B), color='green', label='Class Neg (-1)', marker='x')
plt.title('Original 1D Data')
plt.xlabel('Feature 1')
plt.legend()
plt.grid(True)



plt.tight_layout()
plt.show()
```

```python
# Plot the transformed 2D points (Feature 1 vs. Feature 1 squared)

# Square the 1D points to map them to 2D space
class_A_squared = class_A ** 2
class_B_squared = class_B ** 2


plt.scatter(class_A, class_A_squared, color='red', label='Class Pos (+1)', marker='o')
plt.scatter(class_B, class_B_squared, color='green', label='Class Neg (-1)', marker='x')
plt.title('Transformed 2D Data (Feature 1 vs. Feature 1 Squared)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 1 Squared')
plt.legend()
plt.grid(True)
plt.show()
```
