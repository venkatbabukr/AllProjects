### 1. **Question:**
In **gradient descent**, if the learning rate is set too high, which subtle behavior might be observed?

- A) The gradient descent steps become perfectly aligned with the negative gradient, ensuring guaranteed convergence in fewer epochs.
- B) The cost function values bounce chaotically or even increase over time, failing to converge.
- C) The algorithm accelerates initially but always slows dramatically when near a minimum.
- D) The convergence speed remains unaffected by learning rate adjustments.

<br>
<br>
<br>
<br>
<br>
<br>

**Correct Answer:** B) The cost function values bounce chaotically or even increase over time, failing to converge  
**Explanation:**  
A too-large learning rate can overshoot minima, causing the updates to jump inconsistently and possibly diverge. This behavior appears as oscillations or growing cost values rather than smooth descent.

### 2. **Question:**
When implementing **stochastic gradient descent (SGD)** for a large-scale linear regression problem, which nuanced pitfall might occur?

- A) Batching the data in extremely small batches always guarantees faster convergence and better minima.
- B) The estimator might exhibit high variance in the weight updates, requiring careful tuning of learning rate and batch size.
- C) Convergence is guaranteed after a finite number of iterations, independent of the dataset size.
- D) The weight updates are identical for all samples, making it impossible to differentiate among them.

<br>
<br>
<br>
<br>
<br>
<br>

**Correct Answer:** B) The estimator might exhibit high variance in the weight updates, requiring careful tuning of learning rate and batch size  
**Explanation:**  
SGD updates are computed from one or a few samples at a time, which can fluctuate significantly compared to full-batch methods. A well-chosen learning rate and batch size can mitigate this variance and improve stability.

### 5. **Question:**

- A) Gradual removal of outliers during gradient descent iterations.
- B) Outliers have negligible impact since linear regression always focuses on the median of residuals.
- C) Large errors from outliers can disproportionately affect the cost function, distorting the regression line.
- D) Outliers get automatically truncated if using standard matrix inversion methods.

<br>
<br>

<br>

**Correct Answer:** C) Large errors from outliers can disproportionately affect the cost function, distorting the regression line

### 12. **Question:**
In **precision/recall** terms, which scenario might produce a deceptively high precision score?

- A) A very low number of predicted positives, all of which happen to be correct.
- B) Predicting all samples as positive, thus ensuring almost no false negatives.
- C) A large dataset with many true positives dominating the confusion matrix.
- D) Balancing false positives and false negatives always leads to elevated precision.

<br>
<br>
<br>
<br>
<br>
<br>

**Correct Answer:** A) A very low number of predicted positives, all of which happen to be correct  
**Explanation:**  
Precision measures $\frac{\text{true positives}}{\text{predicted positives}}$. Predicting very few positives can inflate precision if those few predictions are accurate, yet it may ignore many actual positives (harming recall).

### 14. **Question:**
When interpreting **linear regression** coefficients, which subtle phenomenon can mislead about a variable’s importance?

- A) Standardizing the features eliminates all interpretability of coefficients.
- B) If some features are correlated, the coefficients may shift to absorb shared variance, under- or overestimating a feature’s true effect.
- C) Negative coefficient signs automatically mean there’s an inverse causal relationship.
- D) Large absolute coefficients always indicate the most significant predictors.

<br>
<br>
<br>
<br>
<br>
<br>

**Correct Answer:** B) If some features are correlated, the coefficients may shift to absorb shared variance, under- or overestimating a feature’s true effect  
**Explanation:**  
Multicollinearity can cause coefficient values to fluctuate depending on which correlated variables are included. This complicates interpretation, since the effect attributed to one variable might partially belong to another correlated feature.

```python
from imblearn.over_sampling import SMOTE

# Create an instance of SMOTE
smt = SMOTE()

X_sm, y_sm = smt.fit_resample(X_train, y_train)

y_sm.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

st =  StandardScaler()

X_sm = st.fit_transform(X_sm.values)
X_val_scaled = st.transform(X_val.values)
X_test_scaled = st.transform(X_test.values)

y_sm = y_sm.values[:,0]
y_val = y_val.values[:,0]
y_test = y_test.values[:,0]
```

# Hyper-parameters for K-NN

```python
# import knn from sklearn
from sklearn.neighbors import KNeighborsClassifier
```

```python
Acc = []

# Calculating error for K values between 1 and 50
# Note: 51 is not included in the loop
for i in range(1, 51, 2):
    # print("Trying for i: ", i)
    knn = KNeighborsClassifier(n_neighbors=i,metric='euclidean')
    knn.fit(X_sm, y_sm)

    Acc.append(knn.score(X_val_scaled,y_val))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 26), Acc, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Accuracy of K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
```
