```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('cars24-car-price-clean.csv')
df.head()
```

```python
df = df.sample(100)
```

```python
np.random.seed(42)  # For reproducibility

# Add 5 random features (uniform between 0 and 1)
for i in range(1, 15):
    df[f'random_feature_{i}'] = np.random.rand(len(df))
```

```python
df = df.drop('random_feature_13',axis=1)
```

```python
df.columns
```

```python
features = ['year', 'km_driven', 'mileage', 'engine', 'max_power',
       'age', 'make', 'model', 'Individual', 'Trustmark Dealer', 'Diesel',
       'Electric', 'LPG', 'Petrol', 'Manual', '5', '>5','random_feature_1',
       'random_feature_2', 'random_feature_3', 'random_feature_4',
       'random_feature_5', 'random_feature_6', 'random_feature_7',
       'random_feature_8', 'random_feature_9', 'random_feature_10',
       'random_feature_11', 'random_feature_12', 'random_feature_14']
```

```python
y=df[['selling_price']]
X=df[features]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
y_train = np.array(y_train)
```

# y = w1x + [b]

# ideally we want variables with p value lesser than  0.05 - significant variables

```python
# without random variables, r squared is 0.959  adj rs = 0.952
# with random variables, r squared increased to 0.967  adj rs= 0.951
```

```python
X_sm = sm.add_constant(X_train)  # Statmodels default is without intercept, to add intercept we need to add constant.

model = sm.OLS(y_train, X_sm)
results = model.fit()

# Print the summary statistics of the model
print(results.summary())
```
