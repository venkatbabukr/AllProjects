# **Q3. Update weights.**

Given the 1d input features X and target Y. Use Gradient descent to get the optimal weights for the Linear regression model that uses Mean Squared Error(MSE) with L2 regularization.

* The hypothesis function for Linear regression is $\hat{Y} = wX + b$.

* The cost function = $L(w)= 1/2m(\Sigma_{i=1}^m(\hat{Y}_i - Y_i)^2 + \lambda\Sigma_{j=1}^{d}w_j^2)$

Use the below functions to compute the gradient for a single weight w and bias b :

1. w's gradient = $1/m(\Sigma_{i=1}^m(\hat{Y}_i - Y_i)X_i + \lambda w_j)$

2. b's gradient = $1/m(\Sigma_{i=1}^m(\hat{Y}_i - Y_i))$

m is the number of samples,

d is the number of features,

$\lambda$ here is the regularization parameter.

Complete the function to perform L2 regularization and return new updated weight w and bias b, rounded up to 2-decimal places.

```python
import numpy as np

def L2_reg_update_weights(w, b, X, Y, learning_rate, lambda_value=1):
    """input: w is the coefficient of x in form of an integer.
              b is the constant term, integer format.
              X is a python list of input variables.
              Y is a python list of the output variables.
              learning_rate is in the form of float.
              lambda_value is an integer which is actually the penalty term in l2 regularization.
        output: You are required to return updated m and b upto two decimal points, in the same line.
    """

    X = np.asarray(X)
    Y = np.asarray(Y)

    Y_hat = w * X + b
    n = len(Y)

    w_deriv = np.sum(np.dot(Y_hat-Y,X) + lambda_value*w)/n
    b_deriv = np.sum(Y_hat - Y)/n


    # Update weights
    w -= learning_rate * w_deriv
    b -= learning_rate * b_deriv

    return(round(w, 2), round(b, 2))
```

# **Q5. Lasso Regression**

**Problem Description:**

Given the training data, train a Lasso regression model and print out the coefficients and intercept.

```python
import numpy as np


#import Lasso Regression class from scikit-learn library
from sklearn.linear_model import Lasso

def coefficientsAndIntercept(xTrain, yTrain):
    # convert lists into numpy array
    X_train = np.asarray(xTrain)
    y_train = np.asarray(yTrain)

    # initialize Lasso regression and set the value of alpha equal to 1
    regressor = Lasso(alpha = 1)

    # fit the model
    regressor.fit(X_train, y_train)

    #print intercepts and coefficients rounded off upto 2 decimal digits
    print(np.round(regressor.coef_, 2))
    print(np.round(regressor.intercept_, 2))
```

# **AQ2. Which Regularization?**

Which regularization is used to reduce the over fit problem?

### **Ans:**

Both L1 & L2.

**Explanation:**

* Both L1 and L2 regularization methods are used to reduce the overfit problem
* L1 regularization, also known as L1 norm or Lasso (in regression problems), combats overfitting by shrinking the parameters towards 0. This makes some features obsolete.
* L2 regularization, or the L2 norm, or Ridge (in regression problems), combats overfitting by forcing weights to be small, but not making them exactly 0.

# **AQ3. ElasticNet Regularisation**

Given the training and test data, return the predictions for the test data using Elastic net regularized linear regression, rounded up to two decimal places.

```python
import numpy as np

#import elastic net regularized linear regression model
from sklearn.linear_model import ElasticNet


def prediction(xtrain, ytrain, xtest):

    #intantiate the model
    elnet = ElasticNet()

    #train the model
    elnet.fit(xtrain, ytrain)

    #Calculate the prediction for test data
    prd = elnet.predict(xtest)

    return np.round(prd, 2)
```
