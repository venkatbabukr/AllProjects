# **Q2. Residual Space**

If the residual space gets equally increased for data points with both positive and negative residual values , how is the Residuals Sum of Squares (SSR) affected?

**Note:** $SSR = \Sigma_{i=1}^n (y_i - \hat{y_i})^2$

#### **Ans:**

SSR Increases

**Explanation:**

As $(y_i - \hat{y_i})$ is called a Residual and if we increase it equally for Postive and Negative residual valued datapoints:

* The sum of Square $\Sigma (y_i - \hat{y_i})^2$ will increase.
Hence increasing SSR value

**Example:** if $\hat{y}$ = [2,6,8] and y=[3,5,7]

Then Residuals are:

r1:(3 - 2) = 1

r2:(5 - 6) = -1

r3=(7 - 8) = -1

Therefore SSR = $r_1^2 + r_2^2 + r_3^2$ = 1 + 1 + 1 = 3

Now if r1 = 4,r2 = -3,r3 = -5,

Then SSR = $4^2 + (-3)^2 + (-5)^2$ = 16 + 9 + 25 = 50

# **Q3. Calculate the loss value**

Given the data **x** (x[i] represents the values of independent variables for each ith observation), **y** (y[i] represents the value of the dependent variable for ith observation), and **w** (weights), return the sum of loss values for all the observations in x using the following formula:

$Loss = 1/2n * \Sigma_{i=1}^{n} (y_i - w^t.x_i)^2$

**Note:** In this particular question consider the value of intercept(w0) to be zero. The loss should be rounded off to two decimal places.

```python
import numpy as np
def cost(x, y, w):

  x = np.asarray(x)
  w = np.asarray(w)
  y = np.asarray(y)

  loss = 0

  y_hat = np.dot(x, w)
  loss = np.sum((y - y_hat)**2) / (2 * len(y))
  loss = round(loss, 2)

  return loss

x = [[1.0, 0.0], [1.0, 1.0], [1.0, 2.0], [1.0, 3.0], [1.0, 4.0], [1.0, 5.0], [1.0, 6.0], [1.0, 7.0], [1.0, 8.0], [1.0, 9.0]]
y = [32.69, 30.56, 32.04, 32.27, 35.98, 34.39, 40.25, 39.24, 37.14, 39.38]
w = [30.0, 1.0]

print(f"""
x = {x}
y = {y}
w = {w}

cost(x, y, w) = {cost(x, y, w)}
""")
```
