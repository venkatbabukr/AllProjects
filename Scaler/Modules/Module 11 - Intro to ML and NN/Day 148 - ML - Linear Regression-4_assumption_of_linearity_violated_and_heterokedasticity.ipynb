```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Generate nonlinear (quadratic) data
np.random.seed(1)
X = np.linspace(-5, 5, 100)
y = 0.5 * X**2 + 2 * X + 3 + np.random.normal(0, 2, size=X.shape)

# Step 2: Fit a linear regression model
X_reshaped = X.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshaped, y)
y_pred = model.predict(X_reshaped)

# Step 3: Calculate residuals
residuals = y - y_pred

# Step 4: Create a DataFrame
df = pd.DataFrame({
    'X': X,
    'y_true': y,
    'y_pred': y_pred,
    'residual': residuals
})

# Show first few rows
print(df.head())
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, y, label='True Data')
plt.plot(X, y_pred, color='red', label='Linear Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Feature (X) vs Target (y)')
plt.legend()
```

```python
# Step 5: Plot residuals vs predicted values
plt.scatter(y_pred, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Error")
plt.title("Error vs Predicted Values (Nonlinear Pattern)")
plt.show()
```
