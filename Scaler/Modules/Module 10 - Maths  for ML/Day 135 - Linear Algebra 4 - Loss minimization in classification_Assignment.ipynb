# **Q2. Distance**

Let $w_1$ = [4, 3] and $w_2$ = [16, 12] ; $w_0^1$ = 3 and $w_0^2$ = 7, x is a vector

Now let equations of 2 lines be:

L1: $w_1$.x + $w_0^1$ = 0

L2: $w_2$.x + $w_0^2$ = 0

What is the distance between them?

#### **Ans**

The two lines are of the form:

L1: [4 3]$\binom{x}{y}$ + 3 = 0 => 4x + 3y + 3 = 0

L2: [16 12]$\binom{x}{y}$ + 7 = 0 => 16x + 12y + 7 = 0

Slope of two lines m = -4/3,

Y-intercepts $C_1$ = -1 and $C_2$ = -7/12

Distance between lines = (-7/12 + 1)/$\sqrt{1+(-4/3)^2}$

> Using the formula Distance between two parallel lines = |$C_2$ - $C_1$| / $\sqrt{1 + m^2}$

= (5/12)/$\sqrt{25/9}$
= 5/12*3/5 = 1/4

# **Q6. Separating plane**

If the label of point(1, 2) is -1. Given that we have a separating plane 4x + 3y - 12 = 0. What is the correct update equation for w ?

We will use the perceptron update rule for updating the coefficients of classifier, which says that:
1. If a label 1 datapoint is misclassified in the given, we will update w using w = w+x
2. If a label -1 datapoint is misclassified, we will update w using w = w - x

where w is the coefficient vector and x is the variables vector.

**Note:** A datapoint should be labelled 1 if $w^T$x + $w_0$ > 0 and -1 if $w^T$x + $w_0$ < 0, where w is the vector consisting of coefficients of variables and $w_0$ is the constant

#### **Ans:**

No update required, because 4 ( 1 ) + 3 ( 2 ) - 12 = -2, which is < 0, so, the classification is correct as well!
