# **Q1. Local maxima and minima**

Given below function f(x), find the local maxima and minima in the interval [-2,2].

f(x) = 2$x^3$ - 9$x^2$ + 1

#### **Ans:**

Local maxima is at x = 0 where f(x) = 1

**Explanation:**

df/dx = 6$x^2$ - 9x = 6x(x-3).

For maxima/minima, df/dx = 0

=> 6x(x-3) = 0

=> x = 0 or x = 3

Since x = 3 is outside our interval window of [-2, 2], x = 0 is the only solution.

Further, $d^2$f/dx = 12x - 9

At x = 0, $d^2$f/dx = -9 (-ve), So, x = 0 should be point of maxima!

**Note:** At x = 3, $d^2$f/dx = 12.3 - 9 = 27, so, x = 3 is point of local mimima!

# **Q2. PDF maxima**

The probability density function is : f(x) = $\frac{1}{\sigma\sqrt{2\pi}}$ $e^{-(x-\mu)^2/2\sigma^2}$

where μ = mean and σ = standard deviation

The value of 'x' at which the function has maximum value is?

#### **Ans:**

x = $\mu$. Here f(x) = $\frac{1}{\sigma\sqrt{2\pi}}$ is the maximum value

**Explanation:**

Let z(x) = $-(x-\mu)^2/2\sigma^2$

So, f(x) = $\frac{1}{\sigma\sqrt{2\pi}}$ $e^{z(x)}$

f'(x) = $\frac{1}{\sigma\sqrt{2\pi}}$ $e^{z(x)}$ . z'(x)

z'(x) = $-2(x - \mu)/2\sigma^2$ = $(\mu - x)/\sigma^2$

Substituting, we get

f'(x) = $\frac{1}{\sigma\sqrt{2\pi}}$ $e^{-(x-\mu)^2/2\sigma^2}$ . $(\mu - x)/\sigma^2$

For maxima/minima, f'(x) = 0

=> $\frac{1}{\sigma\sqrt{2\pi}}$ $e^{-(x-\mu)^2/2\sigma^2}$ . $(\mu - x)/\sigma^2$ = 0

$e^{-(x-\mu)^2/2\sigma^2}$ can never be zero,

so, we are left with $(\mu - x)/\sigma^2$ = 0

possible only when x = $\mu$



# **Q3. Constrained Equation**

Given a function f(x,y) defined below. Find the point (x,y) at which it is minimum.

f(x, y) = $x^2$ + $y^2$ + 9 such that 3x + 2y - 1 = 0

#### **Ans:**

x = 3/13, y = 2/13 is the constrained minima point of f(x, y)

**Explanation:**

The constrained function $h(x, y, a)$ becomes:

$h(x, y, a)$ = $x^2 + y^2 + 9 + a(3x + 2y - 1)$

$\nabla_{x, y, a}h$ = $\trinom{\delta{h}/\delta{x}}{\delta{h}/\delta{y}}{\delta{h}/\delta{a}}$

= $\trinom{2x + 3a}{2y + 2a}{3x + 2y - 1}$

Now, at minima/maxima, $\nabla_{x, y, a}h$ = 0

=> $2x + 3a$ = 0, $2y + 2a$ = 0, 3x + 2y - 1 = 0.

Solving these three equations, we get x = 3/13 and y = 2/13

# **Q4. Sigmoid derivative**

Given S(x) is a function used in deep learning for classification.

S(x) = $\frac{1}{1 + e^{-x}}$

S'(x) [The derivative of S(x)] = S(x)(1 - S(x))

Suppose we have a function f(x) = -S'(x). The point at which f(x) has a minima and the value of f(x) at that point is?

#### **Ans:**

x = 0 is the point of minima where f(x) = -1/4

**Explanation:**

f(x) = -S'(x) = -S(x)(1-S(x)) = S(x)(S(x) - 1) = $[S(x)]^2$ - S(x)

Now, for f(x) to have minima or maxima, f'(x) = 0

=> (2S(x) - 1)S'(x) = 0

=> (2S(x) - 1)S(x)(1-S(x)) = 0

This is possible only when 2S(x) - 1 = 0 because both S(x) and (1-S(x)) can't become 0.

=> S(x) = 1/2

=> 1/(1 + $e^{-x}$) = 1/2

=> 1 + $e^{-x}$ = 2

=> $e^{-x}$ = 1

=> x = 0 is the point of minima

And when x = 0, S(x) = 1/2 and f(x) = 1/2(1/2 - 1) = -1/4
