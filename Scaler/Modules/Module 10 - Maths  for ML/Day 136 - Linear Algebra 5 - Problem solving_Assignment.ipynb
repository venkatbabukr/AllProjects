1. If a label 1 datapoint is misclassified, we will update w using w = w + x

2. If a label -1 datapoint is misclassified, we will update w using w = w - x

where w is the coefficient vector and x is the variables vector.

What is the new equation after one update using the perceptron update rule?

**Note:** A classifier predicts label 1 if $w^T$x > 0 and -1 if $w^T$x < 0, where w ([$w_1$, $w_2$, $w_0$]) is the weight vector with w0 as intercept and x = [$x_1$, $x_2$, 1]

# **Q3. Coordinates**

We want to represent the point x = (4, 3, 5) in terms of the two vectors: a = (1, 2, -1) and b = (3, 1, 5). What is the new representation of x considering that the vectors a and b are the axes of this new coordinate system?

**Note:** $\hat{a}$ and $\hat{b}$ are unit vectors in the directions of a and b respectively.

#### **Ans:**

$(5/\sqrt{6})\hat{a} + (40/\sqrt{35})\hat{b}$

**Explanation:**

Here the two vectors a and b are orthogonal because 1×3+2×1−1×5=0, therefore we can simply use the projection of point x on these vectors for representation.

||a|| = $\sqrt{(1^2 + 2^2 + -1^2)}$ = $\sqrt{6}$

||b|| = $\sqrt{(3^2 + 1^2 + 5^2)}$ = $\sqrt{35}$

Projection of x on a = x.a/||a||=(4×1+3×2+5×−1)/$\sqrt{6}$ = 5/$\sqrt{6}$

Projection of x on b = x.b/||b||=(4×3+3×1+5×5)/$\sqrt{35}$ = 40/$\sqrt{35}$

So the representation of x on a and b are $(5/\sqrt{6})\hat{a} + (40/\sqrt{35})\hat{b}$

# **Q4. Hyperplanes**

Calculate the distance between the two hyperplanes:

P1: $x_1$ + 2$x_2$ - $x_3$ = 0

P2: 2$x_1$ + 4$x_2$- 2$x_3$ - 8 = 0


#### **Ans**
4/$\sqrt{6}$

**Explanation:**

Norm of weight vector of P1 ||$\vec{w_1}$|| = $\sqrt{(1^2 + 2^2 + -1^2)}$ = $\sqrt{6}$

Norm of weight vector of P2 ||$\vec{w_2}$|| = $\sqrt{(2^2 + 4^2 + -2^2)}$ = $\sqrt{24}$ = 2$\sqrt{6}$

Since the two norms are proportional, the two hyperplanes are parallel to each other!

Distance of P1 from Origin (0, 0, 0) = 0

Distance of P2 from Origin (0, 0, 0) = 8/(2$\sqrt{6}$) = 4/$\sqrt{6}$

Distance between the planes = 4/$\sqrt{6}$ - 0 = 4/$\sqrt{6}$

# **Q6. Positive half space**

Given a plane: 4$x_1$ + 3$x_2$ - 5 = 0, find the equation of a parallel plane that is at a distance of 3 away from this plane and lies in the positive half-space of this plane.

4$x_1$ + 3$x_2$ - 20 = 0

**Explanation:**

* The parallel plane will have the same normal vector as the given plane. Therefore, its equation will be: 4$x_1$ + 3$x_2$ + c = 0, where c is a constant to be determined.

* The distance d between two parallel planes a$x_1$ + b$x_2$ + c1 = 0 and a$x_1$ + b$x_2$ + c2 = 0 is given by the formula: d = |c2 - c1| / $\sqrt{a^2 + b^2}$

* In our case:
  * a=4, b=3
  * The original plane has c1 = -5
  * The parallel plane has c2 = c
  * The required distance is d = 3
* Substituting into the formula:
  * 3 = |c + 5|/$\sqrt{4^2 + 3^2}$
  * => |c + 5| = 15
  * => So, c = 10 or c = -20
* Now, we determine which of these corresponds to the plane lying in the positive half-space of the original plane.
* The positive half-space is defined by: 4$x_1$ + 3$x_2$ - 5 > 0. Among the two values:
  * c = -20 corresponds to the plane 4x1 + 3x2 - 20 = 0, which lies in the positive half-space.
  * c = 10 corresponds to a plane in the negative half-space.
* Therefore, the required parallel plane is: 4$x_1$ + 3$x_2$ - 20 = 0

4. x = -1, y = 5
