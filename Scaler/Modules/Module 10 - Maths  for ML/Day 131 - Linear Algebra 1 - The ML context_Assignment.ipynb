# **Q1. Angle of line**

Suppose you are given an equation of a line y = $1/\sqrt{3}$x + 2. What is the angle of the line w.r.t x-axis?

#### **Ans:**

$30^0$ or $\pi$/6 rad

**Explanation:**

m = tan$\theta$ = 1/$\sqrt{3}$

=> $\theta$ = $\pi$/6 or $30^0$

# **Q4. Slope and Intercept**

#### **Ans:**
Slope = -$w_1$/$w_2$
