# **Q4. Maxima**

f(x) = $x^2$ + x - 10 has a maxima at what value of x?

#### **Ans:**
f(x) has a minima and not maxima at x = -1/2

**Explanation:**

df(x)/dx = 2x + 1

df/dx = 0 => x = -1/2

$d^2$f/dx = 2 (+ve value - means df/dx is increasing)

Therefore at x = -1/2 the function has minima.

It can also be validated graphically as y = $x^2$ + x - 10 is a parabola that has it's minima at x = -1/2

# **Q6. Inflection**

What are the locations of possible inflection points of the function

f(x) = $x^3$/3 + $x^2$/2 - 2x

#### **Ans:**

x = -1/2

**Explanation:**
$d^2$f(x)/dx = 2x + 1

At inflection point, $d^2$f/dx = 0

=> 2x + 1 = 0

=> x = -1/2

**Note:** Inflection point is different from maxima (local/global) or minima (local/global). Inflection point is where you straighten your steering when driving through the curve's curvature
