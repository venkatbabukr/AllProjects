# **Q1. Gradient Ascent**

We use gradient descent to minimize a function. Instead, if we wish to maximize the function, all we have to do is add the gradient term instead of subtracting it in the update equation. If our loss function is f(x) and we want to maximize it with respect to x, which of the following is the correct update equation for gradient ascent?

1. $x^{(t+1)} = x^t - \eta * \frac{df(x^t)}{dx}$
2. $x^{(t+1)} = x^t + \eta * \frac{df(x^t)}{dx}$
3. $x^{(t+1)} = -x^t + \eta * \frac{df(x^t)}{dx}$
4. $x^{(t+1)} = -x^t - \eta * \frac{df(x^t)}{dx}$

#### **Ans:**

Options 2: $x^{(t+1)} = x^t + \eta * \frac{df(x^t)}{dx}$
