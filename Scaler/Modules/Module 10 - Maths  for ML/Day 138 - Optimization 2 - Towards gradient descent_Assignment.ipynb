# **Q6. Apache helicopter**

An Apache helicopter of the enemy is flying along the curve given by y = $x^2$ + 7.

A soldier, placed at (3,7), wants to shoot down the helicopter when it is nearest to him.

What is the nearest distance between the soldier and the helicopter ?

#### **Ans:**

$\sqrt{5}$

**Explanation:**

We need to find point ($x_1$, $y_1$) on the curve that is at a perpendicular distance from (3, 7).

Slope of the curve = dy/dx = 2x

=> Slope of line = -1/2x (As the two lines the tangent to curve and this line are perpendicular to each other, so, $m_c$.$m_l$ = -1)

(y - 7)/(x - 3) = -1/2x

Substitute y = $x^2$ + 7 in the above, we get

$x^2$ / (x - 3) = -1/2x

=> $x^2$ / (3 - x) = 1/2x

=> 2$x^3$ = 3 - x or 2$x^3$ + x - 3 = 0

This equation has only real root at x = 1, which means y = 8 ($1^2$ + 7)

So, the point (1, 8) is the closest point on curve to (3, 7) and distance between them is $\sqrt{(3 - 1)^2 + (7-8)^2}$ = $\sqrt{5}$

# **AQ1. Minimum value**

Find the minimum value of the function $x^2$ - x + 2

#### **Ans:**
7/4

**Explanation:**

dy/dx = 2x - 1

At the point of minimum dy/dx = 0

=> 2x - 1 = 0

=> x = 1/2

At x = 1/2, y = $(1/2)^2$ - 1/2 + 2 = 7/4

# **AQ3. Differentiable functions**

Assume that h(x)=f(g(x)), where both f and g are differentiable functions.

If g(-1)=2, g'(-1)=3, and f'(2)=-4

what is the value of h'(-1)?

#### **Ans:**

-12

**Explanation:**

h'(x) = f'(g(x))*g'(x)

h'(-1) = f'(g(-1))*g'(-1)

=> h'(-1) = f'(2) * 3

=> h'(-1) = -4 * 3 = -12

# **AQ4. Maximum Value**

Find the maximum value of the function: y = logx / x in the domain 0 < x < $\infty$

#### **Ans:**

log e / e

**Explanation:**

dy/dx = -logx/$x^2$ + 1/$x^2$ = 1/$x^2$ (1 - logx)

At maxima/minima dy/dx = 0

=> 1/$x^2$ (1 - logx) = 0

=> logx = 1

=> x = e

and the maximum value y = log e / e
