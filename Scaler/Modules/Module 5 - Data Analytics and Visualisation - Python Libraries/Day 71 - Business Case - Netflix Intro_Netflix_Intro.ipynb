```python
# Movie and TV Show
df_movie = df[df['type'] == 'Movie']
df_movie.head()
```

```python
q3 = df_movie['duration_int'].quantile(0.75)
q1 = df_movie['duration_int'].quantile(0.25)

iqr = q3-q1

ub = q3 + 1.5*iqr
lb = q1 - 1.5*iqr


df_movie[(df_movie['duration_int'] > lb) & (df_movie['duration_int'] < ub)]['duration_int'].describe()
```
