#Sorting

##Sorting a 1D array

```python
#Sorting a 1D array

array_1 = np.array([15,18,7,24,-21,22,8,-10,11])


#np.sort()

print(array_1)
print()
print(np.sort(array_1))
print()
print(array_1)
```

```python
#arr_obj.sort()


array_2 = np.array([15,18,7,24,-21,22,8,-10,11])


print(array_2)
print()
print(array_2.sort())
print()
print(array_2)

```

```python
#Sorting in DESC order

#Method 1

print(array_1)
print()
print(np.sort(array_1)[::-1]) #reversing the result
```

```python
#Another method of sorting

print(array_1)
print()
print(-array_1)
print()
print(np.sort(-array_1))
print()
print(-np.sort(-array_1))
```

##Sorting a 2D array

```python
#2D array

lst_2d = [
    [5,    -7,   46],
    [19,   88,   81],
    [-10, -50,    0],
    [1,     2,    3]
]

arr_2d = np.array(lst_2d)

print(arr_2d)
```

```python
#We will do a simple sort of the above 2d array

print(np.sort(arr_2d))

#observation: For 2d arrays, the default axis = 1

#strictly speaking: the default axis = -1
```

```python
#column wise sorting, along axis = 0

print(np.sort(arr_2d, axis=0))
```

```python
#sort the above 2d array along axis = 1

```

```python
#sort the above 2d array along axis = 0

```

```python
#axis = None
#It first flattens the 2d array, and then sorts the flattened array

print(np.sort(arr_2d, axis=None))
```

```python
shape_of_array = arr_2d.shape

flattened_arr_2d = arr_2d.flatten()

sorted_flattened_arr_2d = np.sort(flattened_arr_2d)

print(sorted_flattened_arr_2d.reshape(shape_of_array))
```

```python
#Custom fuction that takes a number
#and rounds it to the nearest integer

def round_to_integer(x):

  if x >= 0:  #if the input is non-negative

              decimal_part_of_x = x - int(x)

              if decimal_part_of_x >= 0.5:
                return int(x)+1

              else:
                return int(x)

  elif x < 0: #if the input is negative

              decimal_part_of_x = -(x - int(x))

              if decimal_part_of_x >= 0.5:
                  return int(x)-1

              else:
                return int(x)
```

```python
#Testing the round_to_integer() function

x1 = 32.49
print(f"{x1} after rounding: {round_to_integer(x1)}")
print()

x2 = -5.6
print(f"{x2} after rounding:  {round_to_integer(x2)}")
print()


x3 = np.array([98.6, 98.8, 98.1, 98.4])
print(f"{x3} after rounding: {round_to_integer(x3)}")
print()
```

Fahrenheit to Celcius

```python
#Custom fuction that takes a temp value in deg_Fahrenheit
#and returns the corresponding value in deg_Celcius

def fahrenheit_to_celcius(temp_F):
  return (temp_F - 32) * (5/9)
```

```python
#Testing the fahrenheit_to_celcius() function

t1 = 32
print(f"{t1} deg_F = {fahrenheit_to_celcius(t1)} deg_C")
print()

t2 = -5.6
print(f"{t2} deg_F = {fahrenheit_to_celcius(t2)} deg_C")
print()

t3 = np.array([98.6, 98.8, 98.1, 98.4])
print(f"{t3} deg_F = {fahrenheit_to_celcius(t3)} deg_C")
print()
```
