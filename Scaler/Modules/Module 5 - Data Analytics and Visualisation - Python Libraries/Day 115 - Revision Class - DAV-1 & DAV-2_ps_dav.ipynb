```python
You have a DataFrame containing sales transactions. You want to group by "Region"
and compute the standard deviation of sales.
However, you also need a custom transformation where each value
is scaled based on the group's max sale.

Which of the following correctly implements both?

A)df.groupby("Region")["Sales"].std().apply(lambda x: x / x.max())
B)df.groupby("Region")["Sales"].transform(lambda x: x / x.max())
C)df.groupby("Region").apply(lambda x: x["Sales"].std() / x["Sales"].max())
D)df.groupby("Region")["Sales"].agg(["std", lambda x: x / x.max()])
```

```python
import pandas as pd

# Sample data
data = {
    "Region": ["East", "East", "West", "West", "South", "South"],
    "Sales":  [100, 200, 150, 300, 400, 500]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.groupby("Region")["Sales"].std()
```

```python
df.groupby("Region")["Sales"].transform(lambda x: x / x.max())
```

```python
df.groupby("Region").apply(lambda x: x["Sales"].std() / x["Sales"].max())
```
