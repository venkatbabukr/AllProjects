# **Q1. Git Pick**

Which of the following is NOT TRUE about git?

1. By default Git has a master branch
2. git clone operation creates an instance of the repository.
3. git pull operation copies the changes from a local repository to remote repository
4. All of the above

#### **Ans:**

Option 3: git pull operation copies the changes from a local repository to remote repository

# **Q2. About Git Commit**

Which of these statements are TRUE about `commit` in git?

1. A checkpoint of the branch is created and saved in the remote repository.
2. A checkpoint of the branch is created and saved in the local repository.
3. Commits are cheap and should be done frequently.
4. Commits are expensive and should be done only when necessary.

#### **Ans:**

Options 2 & 3:

* A checkpoint of the branch is created and saved in the local repository.
* Commits are cheap and should be done frequently.

# **Q3. Merge Conflict**

When does a merge conflict arise?

**Options:**

1. When two people have changed the same lines in a file.
2. If one developer deleted a file while another developer was modifying it.
3. When git cannot automatically determine what is correct.
4. All of the above

#### **Ans:**

Option 4: All of the above

**Explanation:**: When two people changed same line in a file and git can't determine how to automatically merge the changes, or when one person deletes a file modified by another person and git can't automatically identify whether to keep or remove the file!

# **Q4. Command Not Valid**

Given that 'feature1' is a branch in our project named 'my_project', which of the following is **not a valid** command?

1. git switch feature1
2. git checkout feature1
3. git fork my_project
4. git push origin main

#### **Ans:**

Option 3: git fork my_project. You can't fork using a project name. You can fork from a git repository/repo URL.

# **Q5. About Git Pull**

Which of these statements are CORRECT about `pull` in git?

1. It fetches content from a remote repository.
2. git fetch is safer than git pull because it doesn't alter the local repository.
3. It updates the local repository to match the content of the remote repository.
4. It can be replaced with clone command.

#### **Ans:**

Options 1 - 3:
* It fetches content from a remote repository.
* git fetch is safer than git pull because it doesn't alter the local repository.
* It updates the local repository to match the content of the remote repository.

# **AQ1. New Branch**

Suppose you are working on the main branch of a project and you decide to add a new feature. You decide to create a new branch 'feature_n' for this purpose. In what sequence would you perform the following steps?


1. git commit -m 'commit message'
2. git branch feature_n
3. Make changes to the files
4. git add --all
5. git checkout feature_n

#### **Ans:**

2 -> 5 -> 3 -> 4 -> 1
