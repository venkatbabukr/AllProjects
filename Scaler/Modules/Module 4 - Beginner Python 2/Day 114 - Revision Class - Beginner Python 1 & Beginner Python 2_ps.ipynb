```python
You are developing a banking system that determines a user’s
transaction approval based on the following rules:

If the user is verified (verified == True):`
If the account balance is greater than or equal to the withdrawal amount, approve the transaction.
Otherwise, deny the transaction due to insufficient balance.
If the user is not verified, deny the transaction.
Write a Python function process_transaction(verified, balance, withdraw_amount) that returns:
"Transaction Approved" if conditions are met.
"Transaction Denied: Insufficient Balance" if the balance is low.
"Transaction Denied: User Not Verified" if the user is not verified.
```

```python
def process_transaction(verified, balance, withdraw_amount):
  if verified:  # Check if the user is verified
     if balance >= withdraw_amount:  # Check if balance is enough
        return "Transaction Approved"
     else:
        return "Transaction Denied: Insufficient Balance"
  else:
     return "Transaction Denied: User Not Verified"
```

```python
print(process_transaction(True, 1000, 500))   # Expected Output: Transaction Approved
print(process_transaction(True, 300, 500))    # Expected Output: Transaction Denied: Insufficient Balance
print(process_transaction(False, 1000, 500))  # Expected Output: Transaction Denied: User Not Verified
```

```python
Question 9: You are creating a custom multiplication
table generator for students.

The table should:

Print numbers in a 5x5 format (1 to 5).
Highlight multiples of 3 by surrounding them with [*].
Skip printing multiples of 4 using continue.
Example Output for n=5:
1 2 [*3*] 5 [*6*]
7 [*9*] 10 11 13
14 [*15*] 17 [*18*] 19
[*21*] 22 23 25
```

```python
n = 6
ct = 0
for i in range(1,n**2+1):
  if i %4 == 0 :
    continue
  elif i %3 == 0 :
    print("[*"+str(i)+"*]",end=" ")
    ct = ct + 1
  else :
    print(i,end=" ")
    ct = ct + 1

  if ct %5 == 0 :
    # print("----------")
    print()
```
