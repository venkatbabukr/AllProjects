```python
def pretty(func): # function takes input as a function and returns a function
  def inner():
    print("-"*50) # code that runs before
    func()
    print("-"*50) # code that runs after

  return inner
```

```python
new_func = pretty(foo)

new_func()
```

## Convert the given heights from a list into to t-shirt sizes.
- h < 150 → S
- h >= 150 and h < 180 → M
- h >= 180 → L

```python
def complex_logic(height):
  if height < 150:
      return "S"
  elif height >= 150 and height < 180:
      return "M"
  else:
      return "L"
```

```python
heights = [144, 167, 189, 170, 190, 150, 165, 178, 200, 130]
```

```python
sizes = list(map(complex_logic, heights))
sizes
```

```python
def sum_number(x, y):
    return x + y
```

```python
def summation(x,y,*rajnikant):  #summation(1,2,3,4) [3,4]
  result = x + y
  if rajnikant:
    result += sum(rajnikant)
  return result
```

```python
print(summation(1,2))
print(summation(1,2,3))
print(summation(1,2,3,4))
print(summation(1))
```

```python
def create_person(name, age, gender):
    Person = {
        "name": name,
        "age": age,
        "gender": gender
    }

    return Person

create_person(name = "Rohit", age = 1500, gender = "Male")
```

```python
def create_person(name, age, gender, **extra_info):
    Person = {
        "name": name,
        "age": age,
        "gender": gender
    }

    if extra_info:
        Person.update(extra_info)

    return Person

create_person(name = "Rohit", age = 1500, gender = "Male", color = "blue", hobby = "chess")
```
