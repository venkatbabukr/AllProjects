```python
class Dog:
  kind = "canine" # class variable

  def __init__(self, name):
    self.name = name
```

```python
print(d1.name, d2.name)
```

```python
print(d1.kind, d2.kind)
```

```python
class Student:
    name = "Bipin" # class variable
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def hello(self):
      print(f"Hello {self.name}")

s3 = Student("Ayush", 20)
s3.hello() # if instance and class variables have the same name, the instance variable will be called out by the object
```
