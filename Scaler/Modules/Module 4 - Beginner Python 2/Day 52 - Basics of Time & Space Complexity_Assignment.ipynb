# **Q1. Time complexity of code - I**

What is the time complexity of the following code snippet?

```python
for i in range(n):
  j=i
  while j<i:
    print(j)
    j=j+1
```

#### **Ans:**

O(n)

**Explanation:**

Observe - eventhough there are two loops, the second while loop is doing essentially same as outside for loop.

# **Q2. Triple Nested TC**

What would be the time complexity of the following code?

```python
for i in range(1, n+1):
    for j in range(1, n+1):
        for k in range(n//2, n+1, n//2):
            c += 1
```

#### **Ans:**

The third nested for loop always executes constant times - only 2 times. Other two for loops each execute n times, so, order is $n^2$

# **Q3. isprime() function**

Choose the correct option which represents the time complexity of the function isprime().

```python
def isprime(n):
  for i in range(2, int(n**0.5)+1):
    if n%i==0:
      return False
  return True
```

**Options:**
1. $O(n)$
2. $O(\sqrt{n})$
3. $O(n^2)$
4. $O(log(n))$

#### **Ans:**

Option 2: $O(\sqrt{n})$

# **Q4. Nested For Loops**

For the code given below, what would be the time complexity?

```python
for i in range(1, n):
    for j in range(1, n//4):
        for k in range(1, n):
            break
````

#### **Ans:**

$O(n^2)$. The total number of loop executions = n*n/4 = $n^2$/4 = $O(n^2)$

# **Q5. Complexity of loops**

What is the Big O space complexity of the following piece of code in the terms of n (size of arr)?

```python
def function(arr):
    n = len(arr)
    hash = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            hash[i][j] = arr[i]*2
    return sum([sum(i) for i in hash])

function([1,2,3,4,5])
```

#### **Ans:**

$O(n^2)$. It is creating an nxn matrix from a vector of length n.

# **AQ1. Time complexity of code - III**

What is the time complexity of the following code snippet?

```python
for i in range(n):
  j=i
  while j!=0:
    print(j)
    j=j//2
```

#### **Ans:**

O(nlogn). The inner for loop executes for i + i/2 + i/4 + ... + 1 times = log(i) times for 0 <= i < n
