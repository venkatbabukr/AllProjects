```python
class Car:
  def __init__(self, name, mileage):
    self.name = name
    self.mileage = mileage

  def __str__(self):
    return f"{self.name} -> {self.mileage}"

  def __add__(self, other): # other is an object of class car
    return self.mileage + other.mileage

  def __lt__(self, other):
    return self.mileage < other.mileage

  def __eq__(self, other):
    return self.mileage == other.mileage
```

```python
c1 = Car("Nexon", 12)
c2 = Car("Altroz", 12)
```

```python
print(c1)
```

```python
c1 + c2
```

```python
c1 < c2
# __lt__(c1,c2)
```

```python
c1 > c2
```

```python
c1 == c2
```

```python
def count_char(string):
  hash = {}
  for ch in string[::-1]:
    if hash.get(ch, None) != None:
      hash[ch] += 1
    else:
      hash[ch] = 1

  least_freq = (None, float("inf"))

  for ch, freq in hash.items():
    if hash[ch] < least_freq[1]:
      least_freq = (ch, hash[ch])

  return least_freq[0]
```
