### Metacharacters

Characters having special meaning in RegEx

```
. ^ $ * + ? { } [ ] \ | ( )
```

### Shorthand character classes

* \b: Word boundary
* \d: Number
* \s: White spaces (Space, Tab, Return, New line)
* \w: Alphanumeric

All capital versions (\B, \D, \S, \W) of the above negate their meanings!

### The "Selector" Cheat Sheet

Think of the opening parenthesis and the question mark as a command prompt:

( : "I'm a group."

(? : "I'm a special group... which kind?"

(?= : "...a Lookahead."

(?< : "...a Lookbehind."

(?:: : "...a Non-capturing group."

(?> : "...an Atomic group."

#### Repeating selections

**Back reference**
Because a capturing group () saves text to memory, you can tell the regex to look for that exact same text again later in the string using a backslash and the group number.

***Example:*** (\w+)\s\1
