Swapping

```python
a = 10
b = 30
```

```python
# additional memory method
temp = a
a = b
b = temp
```

```python
print(a,b)
```

```python
# without additional memory 40
a = 10
b = 30

a = a + b
b = a - b # a + b - b
a = a - b

print(a,b)
```
