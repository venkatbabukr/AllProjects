```python
# given a string, count the occurence of each letter
s = "rrsssstttt"

x = {}

for i in s:
  if i in x:
    x[i] += 1
  else:
    x[i] = 1

print(x)
```

```python
s = "rrssssttttrrrbbbkkkllll"
dictionary ={}
for i in s:
  dictionary[i] = s.count(i) # less optimisised. less efficient
dictionary
```

```python
x = (1,2,3,1,2,3,4,4,5,6,6,6,6)

x_unique = list(set(x))
counter_list = [0] * (max(x_unique) + 1)
for i in x:
  counter_list[i] += 1

for i in x_unique:
  print(i, counter_list[i])
```
