# input: hour of the day

1. 9 to 12 : good morning
2. 12 to 17: good afternoon
3. 5 to 24: good night
4. else: please sleep

```python
time = int(input())
if 9<=time<12:
  print("Good morning")
else:
  if time >= 12 and time < 17:
    print("Good afternoon")
  else:
    if time >= 17 and time < 24:
      print("Good night")
    else:
      print("please sleep")
print("execution finished")
```
