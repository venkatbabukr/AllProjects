```python
#Take an input from the user for start and end values. Print all even numbers between them.
start = int(input("start "))
end = int(input("end "))
loop_count = 0
while start<=end:
  loop_count += 1
  if start % 2 == 0:
    print(start)
  start = start + 1

print("no of loops: ",loop_count)
```

```python
# faster code

start = int(input("start "))
end = int(input("end "))
loop_count = 0
while start<=end:
  loop_count += 1
  if start % 2 == 0:
    print(start)
    start+=2
  else:
    start = start + 1

print("no of loops: ",loop_count)
```

```python
# Quiz
x = 1
while x <= 100:
    if x % 3 == 0 and x % 5 == 0:
        print("FizzBuzz")
    elif x % 3 == 0:
        print("Fizz")
    elif x % 5 == 0:
        print("Buzz")
    else:
        print(x)
    x += 1
```

```python
# write code to print table of a number(take number as input)
test_cases = int(input())
t = 1
while t <= test_cases: # try out loop using for instead of while
    number = int(input("Please enter the number: "))
    i = 1
    print("Printing multiplication table of ", number)
    while i <= 10:
        print(number*i, end=" ")
        i += 1
    print("\n")
    t += 1
```

```python
#Calculate sum of all numbers starting from 1 to N using a for loop
number = int(input("Enter a number: "))
total = 0
for value in range(1,number+1):
  total = total + value
print(total)
```
