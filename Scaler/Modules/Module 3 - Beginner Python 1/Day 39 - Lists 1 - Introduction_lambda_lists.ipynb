```python
# find sum, average, min and max

sum_runs = 0
for i in runs_virat:
  sum_runs+=i
print(sum_runs)
```

```python
# find length of the given list
len_runs = 0
for i in runs_virat:
  len_runs+=1
print(len_runs)
```

```python
sum_runs/len_runs
```

```python
# find the max value in the list
max_val = 0
for i in runs_virat:
  if max_val < i:
    max_val = i

print(max_val)
```
