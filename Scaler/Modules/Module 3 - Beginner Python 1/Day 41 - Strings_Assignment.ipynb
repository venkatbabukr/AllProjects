**Q1. Date formatting**

**Problem Description:**

Given the input date in **DD/MM/YYYY** format as a string, write a function that returns the converted date in the string formats **MM/DD/YYYY** and **YYYY/MM/DD** and return the new date formats.

````python
def date_format(date):
    '''Input: date takes the string as input
       Output:return resultant date formats'''

    # YOUR CODE GOES HERE
    dd, mm, yyyy = date.split('/')
    return f"{mm}/{dd}/{yyyy}", f"{yyyy}/{mm}/{dd}"

date = "08/12/1998"
print(f"date_format('{date}') = {date_format(date)}")
````

**Q2. Replace Character**

**Problem Description**

You are given a character string **A** and two integer ASCII codes **B** and **C**.

You have to find and replace all the occurrences of the character having ASCII code equal to **B** with character having ASCII code equal to **C** and return the resultant string.

````python
class Solution:

  def solve(self, A, B, C):
    return A.replace(chr(B), chr(C))

solver = Solution()

A = "aabbccbb"
B = 98
C = 100
print(f"Solution().solve('{A}', {B}, {C}) = {solver.solve(A, B, C)}")
A = "abc"
B = 100
C = 97
print(f"Solution().solve('{A}', {B}, {C}) = {solver.solve(A, B, C)}")
````

**Q4. Reverse the word - II**

**Problem Description**

You are given lowercase string (A) and you have to return after reversing that.

````python
class Solution:
    def solve(self, A):
        return A[::-1]

solver = Solution()

A = "interviewbit"
print(f"Solution().solve('{A}') = {solver.solve(A)}")
A = "scaler"
print(f"Solution().solve('{A}') = {solver.solve(A)}")
````

**Q5. Vowels vs Consonants - II**

**Problem Description**

You are given lowercase string (A) and you have to tell the count of vowels and consonants in it.

````python
VOWELS = "aeiou"

class Solution:
    def solve(self, A):
      alower = A.lower()
      vowel_letters = [letter for letter in alower if letter in VOWELS]
      return [len(vowel_letters), len(A) - len(vowel_letters)]

solver = Solution()

A = "interviewbit"
print(f"Solution().solve('{A}') = {solver.solve(A)}")
A = "scaler"
print(f"Solution().solve('{A}') = {solver.solve(A)}")
````

**AQ6. Palindrome or not**

Write a program to check whether the given string is a palindrome or not. Return True if it is palindrome else return False.

**Note:** A string is said to be a palindrome if the reverse of the string is the same as the string itself.

````python
def isPalindrome(s, ignore_case = False):
    '''Input: s takes the string as input
       Output:return the True or False'''
    # YOUR CODE GOES HERE
    # Optional: check and convert the string to lower case if ignore case
    if ignore_case:
        s = s.lower()
    return s == s[::-1]

s = "madam"
print(f"isPalindrome('{s}') = {isPalindrome(s)}")
s = "Level"
print(f"isPalindrome('{s}', True) = {isPalindrome(s, True)}")
s = "hello"
print(f"isPalindrome('{s}') = {isPalindrome(s)}")
````
