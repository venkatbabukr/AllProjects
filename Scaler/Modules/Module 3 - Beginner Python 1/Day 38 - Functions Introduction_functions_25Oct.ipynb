```python
def do_something(n):
    total = 0
    i = 1
    while i <= n:
        total += i
        i += 1
    return total

print(do_something(5))
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

x = gcd(48, 180)
print(x)
```

```python
#keyword based argument
def introduce_family(my_name, sibling_name, father_name, mother_name):
    print("My name is", my_name)
    print("My sibling's is", sibling_name)
    print("My father's is", father_name)
    print("My mother's is", mother_name)

# Giving arguments in incorrect order
introduce_family("Nilesh", "Deepak", "Deepa", "Vidhi")
```

```python
introduce_family(father_name="Deepak",sibling_name="Vidhi",my_name="Nilesh",mother_name="Deepa")
```

```python
#default argument values

def simple_interest(p,t,r=5): # default values should be at the end
  interest = (p * r * t) / 100
  return interest
```

```python
print(simple_interest(7500,5))
```

```python
print(simple_interest(7500,2))
```

```python
print(simple_interest(7500,2,10))
```
