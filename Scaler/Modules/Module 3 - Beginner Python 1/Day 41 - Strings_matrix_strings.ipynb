```python
n = int(input())
outer = []
for i in range(n):
  inner = []
  for j in range(n):
    inner.append(int(input()))
  outer.append(inner)
```

```python
outer
```

```python
# universal approach
sum = 0
for i in range(len(outer)): # indices
  for j in range(len(outer[i])): # indices
    sum+=outer[i][j]
print(sum)
```

```python
# pythonic approach
sum = 0
for i in outer:
  for j in i:
    sum+=j
print(sum)
```

### check palindrome

```python
def check_palindrome(a):
  b = a.lower()
  return b == b[::-1]
```

```python
check_palindrome("Abba")
```
