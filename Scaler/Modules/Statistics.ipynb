## Harmonic Mean

$$HM(a, b) = \frac{2}{\frac{1}{a} + \frac{1}{b}} = \frac{2ab}{a + b}$$

### Uses
Arithmetic mean can easily be pulled up by few high value outliers. Harmonic mean can be used in following cases:

1. When we need to respect low values during calculation.  
**Examples:**  
A1 = [1, 2, 3, 4, 5, 5000]  
***AM(A1) ~= 836,   
HM(A1) ~= 2.63*** (Shows the data still has significantly smaller numbers)

2. When we want both numbers a and b to be high or low, taking harmonic mean will help us.
    * Arithmetic mean suffers from divergence. When first number is extremely high (95%) and second number is extremely low (5%), arithmetic mean = 50% will give a wrong picture. However, harmonic mean = 9.5% gives us a much clearer picture.  
**Example application:** In Logistic regression, whenever we want both ***precision*** and ***recall/sensitivity*** to be high, we take harmonic mean of those two.

### Relationship with Arithmetic mean

$$\boxed{AM(a, b) * HM(a, b) = ab}$$

If Arithmetic mean can be considered as LCM, then Harmonic mean will be like GCD. The structural relationship is 100% accurate, but with following caveats:

The functional definitions differ:
1. ***Linear vs. Divisibility:*** The Arithmetic Mean is based on addition (a + b), whereas the LCM is based on prime factors (multiplication).
2. ***Scaling:*** If you double a and b, the AM and HM both exactly double. However, the GCD and LCM behave differently based on the shared factors of the new numbers.
