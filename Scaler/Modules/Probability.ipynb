# Conditional Probability

## Law of Total Probability

**Intution/Understanding**

When the total samplespace is divided into n partitions by events $B_1, B_2, B_3 ... B_n$, then Probability of an overlapping event A is given by formula:

$\boxed{P(A) = P(A|B_1).P(B_1) + P(A|B_2).P(B_2) + P(A|B_3).P(B_3) + ... + P(A|B_n).P(B_n)}$

**Uses**

Use this whenever you wish to calculate P(A) from P(A|B) and P(A|B') using formula:

$\boxed{P(A) = P(A|B).P(B) + P(A|B').P(B')}$

## Baye's Theorem

**Intuition/Understanding**

It gives us ***correct factor*** by which P(an event E1) will ***scale-up or scale-down*** in presence of another event E2...

$\boxed{P(E1 | E2) = P(E1).\frac{P(E2|E1)}{P(E2)} }$

**Proof**

P(E1 | E2) = P(E1 $\cap$ E2)/P(E2) => P(E1 $\cap$ E2) = P(E1 | E2).P(E2)

P(E2 | E1) = P(E1 $\cap$ E2)/P(E1) => P(E1 $\cap$ E2) = P(E2 | E1).P(E1)

Therefore,

P(E1 | E2).P(E2) = P(E2 | E1).P(E1)

=> $\boxed{P(E1 | E2) = P(E1).P(E2 | E1)/P(E2)}$

## Independent Events

An event E2 is said to be independent of E1 if it's probability in presence of E1 is same as it's probability overall!

Mathematically, E1 and E2 are independent events if:

$\boxed{P(E2|E1) = P(E2)}$

Combining the above with conditional probability $P(E1 \cap E2) = P(E2|E1).P(E1)$

=> For independent events, $\boxed{P(E1 \cap E2) = P(E2).P(E1)}$

**Intuition/Visualization/Understanding**

* The probability of E2 is not influenced in anyway by E1. i.e. **P(E2) remains same irrespective of whether E1 happens or not**, then E1 & E2 are independent.
    * Any chain of events where **outcome of one provides no new information about the underlying process of the other** are definitely independent.
    * Most common but not true always is - **sequence of events not related by memory**   
    Ex: **Random flip of coins one after another** (Every flip is independent)

#### Corollary 1: Two mutually exclusive events are never independent of each other

**Proof:** Since the two events are mutually exclusive, $P(E1 \cap E2) = 0$ != P(E1).P(E2). Therefore, they can't be independent of each other!

This is intuitive to understand too because if are chained in a sequence of execution and let any one of the event happen, then, the probability of other event will become 0, so, they are not independent.
